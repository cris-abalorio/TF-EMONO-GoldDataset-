# Library

```python
import pandas as pd1
import pandas as pd
data=pd.read_csv('gold-dataset-sinha-khandait-NEWS-SOURCE-MERGED-VERSION.csv')

import nltk
stopwords = nltk.corpus.stopwords.words('english')
import string
import re
from nltk.stem import PorterStemmer
ps = PorterStemmer()
import numpy as np
```


# Text Cleaning (combined)

```python
def clean_text(txt):
    txt = "".join([c for c in txt if c not in string.punctuation])
    tokens = re.split('\W+',txt)
    txt = [ps.stem(word) for word in tokens if word not in stopwords]
    return txt
    
```

# 1. Get all unique terms

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(analyzer=clean_text, min_df=2)
vect.fit(data['document'])
X = vect.transform(data['document'])
```

```python
features = vect.get_feature_names()
```
